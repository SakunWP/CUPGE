Voici la classe BinaryTree vue en cours. Observer la défintion de chaque méthode

````python
class BinaryTree():
    def __init__(self,data,left=None,right=None):
      self.data =data
      self.left =left
      self.right =right
    
    def printTree(self):
      pr.printBTree(self,lambda n:(str(n.data),n.left,n.right)) 
    
    def child_left(self):
        return self.left
    
    def child_right(self):
        return self.right
    
    def racine(self):
        return self.data
    
    def empty_tree(self):
    	return (self.data!=None and self.left==None and self.right==None)
    
````
